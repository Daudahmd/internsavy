```python
final = df.drop(['Serial No.'],axis = 'columns')
final
```

#Check Nan Values

```python
final.isna().sum()
```

#Plot a graph

```python
plt.xlabel('CGPA')
plt.ylabel('Chance of Admit ')
plt.scatter(final.CGPA, final['Chance of Admit '])
```

```python
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit ')
plt.scatter(final['GRE Score'], final['Chance of Admit '],color = 'red')
```

```python
x = final.drop(['Chance of Admit '], axis = 'columns')
y = final['Chance of Admit ']
```

#Import Model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

#Split data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.predict(x_test)
```
